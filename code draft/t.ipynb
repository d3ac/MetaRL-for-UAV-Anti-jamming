```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.linspace(0, 4, 1000)
x2 = np.linspace(0, 4, 1000)
y1 = np.arcsin((x1-2)/2)
y2 = 2 * np.arcsin(np.sqrt(x2)/2)

plt.plot(x1, y1, label='y1 = arcsin((x-2)/2)')
plt.plot(x2, y2, label='y2 = arcsin(sqrt(x)/2)')
plt.legend();
```
