# reward

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib_inline import backend_inline
backend_inline.set_matplotlib_formats('svg')

index_max = 0

data1 = pd.read_csv('DQN/reward.csv').values[:,1]
data2 = pd.read_csv('DQN-context/reward.csv').values[:,1]
data3 = pd.read_csv('DDRQN/reward.csv').values[:,1]
data4 = pd.read_csv('DDRQN-context/reward.csv').values[:,1]
data5 = pd.read_csv('RANDOM/reward.csv').values[:,1]
data6 = pd.read_csv('PPO/reward.csv').values[:,1]
data7 = pd.read_csv('PPO-context/reward.csv').values[:,1]

index_max = min(len(data1), len(data2), len(data3), len(data4), len(data5), len(data6), len(data7))
# index_max = min(len(data1), len(data2), len(data3), len(data5))
data1 = data1[:index_max]
data2 = data2[:index_max]
data3 = data3[:index_max]
data4 = data4[:index_max]
data5 = data5[:index_max]
data6 = data6[:index_max]
data7 = data7[:index_max]

x = np.arange(len(data1))

def moving_average(data, window_size):
    processed_data = []
    for i in range(len(data)):
        if i < window_size:
            processed_data.append(np.mean(data[:i+1]))
        else:
            processed_data.append(np.mean(data[i-window_size:i]))
    return processed_data

ws = 50

data1_moving = moving_average(data1, ws)
data2_moving = moving_average(data2, ws)
data3_moving = moving_average(data3, ws)
data4_moving = moving_average(data4, ws)
data5_moving = moving_average(data5, ws)
data6_moving = moving_average(data6, ws)
data7_moving = moving_average(data7, ws)

plt.plot(x, data1, color='#e75840', alpha=0.2)
plt.plot(x, data1_moving, color='#e75840', label='dqn')

plt.plot(x, data2, color='#4e79a7', alpha=0.2)
plt.plot(x, data2_moving, color='#4e79a7', label='dqn-context')

plt.plot(x, data3, color='#f28e2b', alpha=0.2)
plt.plot(x, data3_moving, color='#f28e2b', label='ddqn')

plt.plot(x, data4, color='#59a14f', alpha=0.2)
plt.plot(x, data4_moving, color='#59a14f', label='ddqn-context')

plt.plot(x, data5, color='#b07aa1', alpha=0.2)
plt.plot(x, data5_moving, color='#b07aa1', label='random')

plt.plot(x, data6, color='#7f6000', alpha=0.2)
plt.plot(x, data6_moving, color='#7f6000', label='PPO')

plt.plot(x, data7, color='#76EE00', alpha=0.2)
plt.plot(x, data7_moving, color='#76EE00', label='PPO-context')

plt.ylim(-900, 300)

plt.legend()
plt.grid(True)
plt.title('reward')
plt.xlabel('time steps')
plt.ylabel('average return')

plt.savefig('reward_plot.svg', format='svg')  # 指定文件路径和名称  
plt.show()
```
